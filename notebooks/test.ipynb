# prime mathematics

prime mathematics is a numbering system based on prime numbers

```python
# number representation
alphabet = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']

# calculate the number system
def prime(n):
    l = 0
    numbering_system = []
    for i in range(2,n):
        for j in range(2,i):
            if i%j==0:
                break
        else:
            numbering_system.append(i)
            l+=1
    
    return numbering_system

# initialize the number system
numbering_system = prime(100)
print ("Numbering system initialized: ", numbering_system)

# convert the number to the number system
def conver_to_prime_system(number):
    code = {}

    j = 0
    while number>1 :
        for j in range(0,len(numbering_system)):
            if number%numbering_system[j]==0:
                code[alphabet[j]] = code.get(alphabet[j], 0)+1
                number = number/numbering_system[j]

    return code

# convert the number system to the number
def convert_to_decimal(prime_number):
    number = 1

    # iterate over the dictionary
    for key, value in prime_number.items():
        number *= numbering_system[alphabet.index(key)]**value
    
    return number

# representation of the number system
def serialize(prime_number):
    serialize = ''
    for key, value in prime_number.items():
        serialize += key + ('' if value ==1 else str(value))

    return serialize

# test
print ("Test: ", convert_to_decimal(conver_to_prime_system(100)) == 100)

# coversion test
for i in range(1,100):
    print (i, serialize(conver_to_prime_system(i)), convert_to_decimal(conver_to_prime_system(i)))
```
